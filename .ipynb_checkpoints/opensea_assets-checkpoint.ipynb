```python
with open('assets.json') as f:
    data = json.loads(f.read())
    asset_rarities=[]

    for asset in data['assets']:
        asset_rarity = 1
        for trait in asset['traits']:
            trait_rarity = trait['trait_count'] / 8888
            asset_rarity += trait_rarity

        if asset['last_sale']:
            price = asset['last_sale']
        else: price = 0
            
        if asset['creator']['user']['username']:
            creator = asset['creator']['user']['username']
        else: creator = '0'
        asset_rarities.append({
                'token_id' : asset['token_id'],
                'creator': creator,
                'description' : asset['description'],
                'rarity' : asset_rarity,
                'traits' : asset['traits'],
                'image_url' : asset['image_url'],
                'collection': asset['collection']['name'],
            'created': asset['asset_contract']['created_date'],
            'price': price,
                     })
        
    assets_sorted = pd.DataFrame(sorted(asset_rarities, key=lambda asset: asset['rarity']))
    assets_sorted = pd.DataFrame(asset_rarities)
           
#for asset in assets_sorted[:20]:
#    render_asset(asset)
#    st.subheader(f"{len(asset['traits'])} Traits")
#    for trait in asset['traits']:
#        st.write(f"{trait['trait_type']} - {trait['value']} - {trait['trait_count']} have this")
```
